```python
# Import thư viện
import pandas as pd
import numpy as np
```

```python
# Đọc dữ liệu CSV
df = pd.read_csv("../data/raw/xau_usd_historical_data.csv")
df.head()
```

```python
# Kiểm tra thông tin dữ liệu
print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nInfo:")
print(df.info())
```

```python
# Đổi tên cột về chuẩn
df = df.rename(columns={
    "Date": "date",
    "Price": "close",
    "Open": "open",
    "High": "high",
    "Low": "low",
    "Vol.": "volume",
    "Change %": "change_pct"
})
df.head()
```

```python
# Chuyển cột date về kiểu datetime
df["date"] = pd.to_datetime(df["date"], format="%Y/%m/%d")
df[["date"]].head()
```

```python
# Chuyển các cột giá về kiểu số thực
price_cols = ["close", "open", "high", "low"]

for col in price_cols:
    df[col] = df[col].astype(str).str.replace(",", "", regex=False)
    df[col] = pd.to_numeric(df[col], errors="coerce")

df[price_cols].head()
```

```python
# Chuyển cột change_pct về kiểu float

df["change_pct"] = (
    df["change_pct"]
    .astype(str)
    .str.strip()                       # bỏ khoảng trắng đầu cuối
    .str.replace('"', '', regex=False) # bỏ dấu nháy kép
    .str.replace("%", "", regex=False) # bỏ dấu %
    .str.replace("−", "-", regex=False) # nếu có dấu âm unicode
)

df["change_pct"] = pd.to_numeric(df["change_pct"], errors="coerce") / 100
df[["change_pct"]].head()
```

```python
# Xem lại thông tin dữ liệu sau khi đã chuyển đổi kiểu dữ liệu
print("\nInfo after type conversion:")
print(df.info())
df.head()
```

```python
# Sắp xếp dữ liệu theo ngày tăng dần

df = df.sort_values("date").reset_index(drop=True)
df.head(), df.tail()
```

```python
# Kiểm tra ngày bị trùng lặp

duplicate_dates = df["date"].duplicated().sum()
print("Số ngày bị trùng:", duplicate_dates)

# df[df["date"].duplicated(keep=False)].sort_values("date") Nếu muốn xem cụ thể các ngày bị trùng
# Xử lý ngày bị trùng (nếu có) bằng cách giữ lại bản ghi đầu tiên và loại bỏ các bản ghi trùng lặp khác
if duplicate_dates > 0:
    df = df.drop_duplicates(subset=["date"]).reset_index(drop=True)
```

```python
# Kiểm tra missing values
df.isnull().sum()
```

```python
# Kiểm tra logic OHLC

invalid_ohlc = df[
    (df["high"] < df["open"]) |
    (df["high"] < df["close"]) |
    (df["low"] > df["open"]) |
    (df["low"] > df["close"]) |
    (df["high"] < df["low"])
]

print("Số dòng OHLC bất thường:", len(invalid_ohlc))
invalid_ohlc.head()

# Nếu Số dòng OHLC bất thường = 0 thì dữ liệu đã sạch về mặt logic OHLC, nếu có thể xem chi tiết các dòng bất thường để xử lý nếu cần thiết
```

```python
# Giữ lại các cột cần thiết cho việc dự báo giá vàng
df_clean = df[["date", "open", "high", "low", "close"]].copy()
df_clean.head()
```

```python
# Kiểm tra lại type dữ liệu trước khi lưu file
print(df_clean.dtypes)
print(df_clean.head())
```

```python
# Lưu dữ liệu đã làm sạch vào file mới
df_clean.to_csv("../data/processed/gold_cleaned.csv", index=False)
print("Đã lưu file: ../data/processed/gold_cleaned.csv")
```
