## Tiền xử lý dữ liệu

```python
# Import thư viện
import pandas as pd
```

```python
# Đọc dữ liệu CSV
df = pd.read_csv("../data/raw/xau_usd_historical_data.csv")
df.head()
```

```python
# Kiểm tra thông tin dữ liệu
print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nInfo:")
print(df.info())
```

```python
# Đổi tên cột về chuẩn
df = df.rename(columns={
    "Date": "date",
    "Price": "close",
    "Open": "open",
    "High": "high",
    "Low": "low",
    "Vol.": "volume",
    "Change %": "change_pct"
})
df.head()
```

```python
# Chuyển cột date về kiểu datetime
df["date"] = pd.to_datetime(df["date"], format="%Y/%m/%d")
df[["date"]].head()
```

```python
# Chuyển các cột giá về kiểu số thực
price_cols = ["close", "open", "high", "low"]

for col in price_cols:
    df[col] = df[col].astype(str).str.replace(",", "", regex=False)
    df[col] = pd.to_numeric(df[col], errors="coerce")

df[price_cols].head()
```

```python
# Chuyển cột change_pct về kiểu float

df["change_pct"] = (
    df["change_pct"]
    .astype(str)
    .str.strip()                       # bỏ khoảng trắng đầu cuối
    .str.replace('"', '', regex=False) # bỏ dấu nháy kép
    .str.replace("%", "", regex=False) # bỏ dấu %
    .str.replace("−", "-", regex=False) # nếu có dấu âm unicode
)

df["change_pct"] = pd.to_numeric(df["change_pct"], errors="coerce") / 100
df[["change_pct"]].head()
```

```python
# Xem lại thông tin dữ liệu sau khi đã chuyển đổi kiểu dữ liệu
print("\nInfo after type conversion:")
print(df.info())
df.head()
```

```python
# Sắp xếp dữ liệu theo ngày tăng dần

df = df.sort_values("date").reset_index(drop=True)
df.head(), df.tail()
```

```python
# Kiểm tra ngày bị trùng lặp

duplicate_dates = df["date"].duplicated().sum()
print("Số ngày bị trùng:", duplicate_dates)

# df[df["date"].duplicated(keep=False)].sort_values("date") Nếu muốn xem cụ thể các ngày bị trùng
# Xử lý ngày bị trùng (nếu có) bằng cách giữ lại bản ghi đầu tiên và loại bỏ các bản ghi trùng lặp khác
if duplicate_dates > 0:
    df = df.drop_duplicates(subset=["date"]).reset_index(drop=True)
```

```python
# Kiểm tra missing values
df.isnull().sum()
```

```python
# Kiểm tra logic OHLC

invalid_ohlc = df[
    (df["high"] < df["open"]) |
    (df["high"] < df["close"]) |
    (df["low"] > df["open"]) |
    (df["low"] > df["close"]) |
    (df["high"] < df["low"])
]

print("Số dòng OHLC bất thường:", len(invalid_ohlc))
invalid_ohlc.head()

# Nếu Số dòng OHLC bất thường = 0 thì dữ liệu đã sạch về mặt logic OHLC, nếu có thể xem chi tiết các dòng bất thường để xử lý nếu cần thiết
```

```python
# Kiểm tra outliers bằng phương pháp IQR cho các cột giá
price_cols = ["open", "high", "low", "close"]

outlier_summary = {}

for col in price_cols:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

    outlier_summary[col] = {
        "Q1": q1,
        "Q3": q3,
        "IQR": iqr,
        "Lower Bound": lower_bound,
        "Upper Bound": upper_bound,
        "Outlier Count": len(outliers)
    }

outlier_count_df = pd.DataFrame(outlier_summary).T
outlier_count_df = outlier_count_df.reset_index().rename(columns={"index": "feature"})
outlier_count_df
```

```python
# Vẽ biểu đồ cột số lượng outliers cho từng cột giá
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
bars = plt.bar(outlier_count_df["feature"], outlier_count_df["Outlier Count"])

plt.title("Số lượng outlier theo từng biến giá")
plt.xlabel("Biến giá")
plt.ylabel("Số lượng outlier")

# Hiển thị số trên đầu cột
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height + 0.05,
        f"{int(height)}",
        ha="center",
        va="bottom"
    )

plt.tight_layout()
plt.show()
```

```python
import pandas as pd


# -------------------------
# Bước 0: thống kê ban đầu
# -------------------------
initial_rows = len(df)

# -------------------------
# Bước 1: loại các dòng sai logic OHLC
# -------------------------
invalid_ohlc_mask = (
    (df["high"] < df["open"]) |
    (df["high"] < df["close"]) |
    (df["low"] > df["open"]) |
    (df["low"] > df["close"]) |
    (df["high"] < df["low"])
)

invalid_ohlc_rows = df[invalid_ohlc_mask].copy()
removed_ohlc_count = len(invalid_ohlc_rows)

df_after_ohlc = df[~invalid_ohlc_mask].copy().reset_index(drop=True)

# -------------------------
# Bước 2: xác định outlier của close bằng IQR
# (thực hiện trên dữ liệu sau khi đã loại OHLC bất thường)
# -------------------------
q1 = df_after_ohlc["close"].quantile(0.25)
q3 = df_after_ohlc["close"].quantile(0.75)
iqr = q3 - q1

lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

close_outlier_mask = (
    (df_after_ohlc["close"] < lower_bound) |
    (df_after_ohlc["close"] > upper_bound)
)

close_outlier_rows = df_after_ohlc[close_outlier_mask].copy()
removed_close_outlier_count = len(close_outlier_rows)

# -------------------------
# Bước 3: loại outlier close
# -------------------------
df_final = df_after_ohlc[~close_outlier_mask].copy().reset_index(drop=True)

final_rows = len(df_final)

# -------------------------
# Bước 4: tổng hợp kết quả
# -------------------------
summary_df = pd.DataFrame({
    "Tiêu chí": [
        "Số dòng ban đầu",
        "Số dòng bị loại do sai logic OHLC",
        "Số dòng còn lại sau khi loại OHLC bất thường",
        "Q1 của close",
        "Q3 của close",
        "IQR",
        "Ngưỡng dưới (Lower Bound)",
        "Ngưỡng trên (Upper Bound)",
        "Số dòng bị loại do outlier của close",
        "Số dòng cuối cùng còn lại"
    ],
    "Giá trị": [
        initial_rows,
        removed_ohlc_count,
        len(df_after_ohlc),
        round(q1, 4),
        round(q3, 4),
        round(iqr, 4),
        round(lower_bound, 4),
        round(upper_bound, 4),
        removed_close_outlier_count,
        final_rows
    ]
})

summary_df

```

```python
# Giữ lại các cột cần thiết cho việc dự báo giá vàng
df_clean = df_final[["date", "open", "high", "low", "close"]].copy()
df_clean.head()
```
